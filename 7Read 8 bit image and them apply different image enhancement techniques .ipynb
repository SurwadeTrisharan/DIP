```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('pout (1).jpg',cv2.IMREAD_GRAYSCALE)
```

```python
bi = cv2.equalizeHist(img) #Brightness Imporvement
```

```python
br = img // 2        #Brightness Reduction
```

```python
bw = cv2.threshold(img, 127,255,cv2.THRESH_BINARY)[1]   #Binary Image(Thresholding at 127)
```

```python
bw2 = img < 110  #Thresholding at 110
```

```python
ni = cv2.bitwise_not(img)  #Negative Image
```

```python
lt = np.log1p(img.astype(float))   #Log Transformation
```

```python
gamma = 2  #Power Law Transformation(Gamma Correction)
a=1
plt_img = a* np.power(img.astype(float),gamma)
```

```python
plt.figure(figsize = (12,12))

plt.subplot(3,3,1)
plt.imshow(img, cmap = 'gray')
plt.title('Original Image')

plt.subplot(3,3,2)
plt.imshow(bi, cmap = 'gray')
plt.title('Brightness Improvement')

plt.subplot(3,3,3)
plt.imshow(br, cmap = 'gray')
plt.title('Brightness Reduction')

plt.subplot(3,3,4)
plt.imshow(bw,cmap = 'gray')
plt.title('Binary Image')

plt.subplot(3,3,5)
plt.imshow(bw2, cmap='gray')
plt.title('Thresholding = 110')

plt.subplot(3,3,5)
plt.imshow(ni,cmap = 'gray')
plt.title('Negative of the Image')

plt.subplot(3,3,6)
plt.imshow(lt,cmap = 'gray')
plt.title('Log Transformation')

plt.subplot(3,3,7)
plt.imshow(plt_img, cmap='gray')
plt.title('Power Law Transformation')

plt.show()
```
