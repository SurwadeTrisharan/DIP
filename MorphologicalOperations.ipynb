```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img1=cv2.imread('Cameraman.png',cv2.IMREAD_GRAYSCALE)
```

```python
_,img=cv2.threshold(img1,128,255,cv2.THRESH_BINARY)
```

```python
se = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(11,11))
```

```python
close_bw = cv2.morphologyEx(img, cv2.MORPH_CLOSE, se)
open_bw = cv2.morphologyEx(img,cv2.MORPH_OPEN,se)
dilate_bw = cv2.dilate(img,se)
erode_bw = cv2.erode(img,se)
```

```python
plt.figure(figsize=(20,20))
```

```python
plt.subplot(3, 3, 1)
plt.imshow(img1, cmap='gray')
plt.title('Original Image')

plt.subplot(3, 3, 2)
plt.imshow(img, cmap='gray')
plt.title('Binary Image')

plt.subplot(3, 3, 3)
plt.imshow(close_bw, cmap='gray')
plt.title('Closing Morphological Operation')

plt.subplot(3, 3, 4)
plt.imshow(open_bw, cmap='gray')
plt.title('Opening Morphological Operation')

plt.subplot(3, 3, 5)
plt.imshow(dilate_bw, cmap='gray')
plt.title('Dilation Morphological Operation')

plt.subplot(3, 3, 6)
plt.imshow(erode_bw, cmap='gray')
plt.title('Erosion Morphological Operation')
plt.show()
```
